## Reading files

Now that the files are renamed, we can now train a model on the training set, which first needs to be loaded in in a suitable fashion so that the model can use the images as inputs.

```python
import os
from sklearn.metrics import confusion_matrix
import seaborn as sn; sn.set(font_scale=1.4)
from sklearn.utils import shuffle           
import matplotlib.pyplot as plt             
import cv2         
from tqdm import tqdm
```

```python
class_names = ['back', 'sideon', 'topdown']
class_names_label = {class_name:i for i, class_name in enumerate(class_names)}

nb_classes = len(class_names)

IMAGE_SIZE = (227, 227)
```

```python
def load_data():
    
    #datasets = [r'C:\Users\shah_\OneDrive\Desktop\_Data Science Toolbox\_Assessment 3\Classification Folder\_Combined\Train', r'C:\Users\shah_\OneDrive\Desktop\_Data Science Toolbox\_Assessment 3\Classification Folder\_Combined\Test']
    datasets = [r'C:\Users\shah_\OneDrive\Desktop\_Data Science Toolbox\_Assessment 3\Classification Folder\_Combined\Train']
    output = [[],[]]
    
    # Iterate through training and test sets
    for dataset in datasets:
        
        images = []
        labels = []
        
        print("Loading {}".format(dataset))
        
        # Iterate through each folder corresponding to a category
        for folder in os.listdir(dataset):
            label = class_names_label[folder]
            
            # Iterate through each image in our folder
            for file in tqdm(os.listdir(os.path.join(dataset, folder))):
                
                # Get the path name of the image
                img_path = os.path.join(os.path.join(dataset, folder), file)
                
                # Open the img
                image = cv2.imread(img_path) 
                
                # Append the image and its corresponding label to the output
                images.append(image)
                labels.append(label)
                
        images = np.array(images, dtype = 'float32')
        labels = np.array(labels, dtype = 'int32')   
        
        #output.append((images, labels))
        output[0].append(images)
        output[1].append(labels)

    return output
```

```python
(train_images_temp, train_labels_temp) = load_data()
train_images = train_images_temp[0]
train_labels = train_labels_temp[0]
```

```python
print(train_labels)
```

```python
n_train = len(train_labels)

print ("Number of training examples: {}".format(n_train))
print ("Each image is of size: {}".format(IMAGE_SIZE))

print("Number of back angle: {}".format(np.count_nonzero(train_labels==0)))
print("Number of sideon angle: {}".format(np.count_nonzero(train_labels==1)))
print("Number of topdown angle: {}".format(np.count_nonzero(train_labels==2)))
```

We now display some scans as a test to check everything has worked.

```python
def display_examples(class_names, images, labels, start):
    """
        Display 25 images from the images array with its corresponding labels
    """
    
    fig = plt.figure(figsize=(10,10))
    fig.suptitle("Some examples of images of the dataset", fontsize=16)
    for i in range(25):
        plt.subplot(5,5,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(images[i + start].astype('uint8'), cmap=plt.cm.binary)
        plt.xlabel(class_names[labels[i + start]])
    plt.show()
```

## Model Training

We now train a neural network model from keras using the loaded images above as inputs.

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (227, 227, 3)), 
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(32, (3, 3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(6, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_images, train_labels, batch_size=128, epochs=12, validation_split = 0.2)
```

## Prediction

We now load the test set.

```python
def load_test():
    
    
    folder = r'C:\Users\shah_\OneDrive\Desktop\_Data Science Toolbox\_Assessment 3\Classification Folder\_Combined\Test'
    output = []
    
    images = []
    
    # Iterate through each image in our folder
    for file in tqdm(os.listdir(folder)):
                
        # Get the path name of the image
        img_path = os.path.join(os.path.join(folder), file)
                
        # Open the img
        image = cv2.imread(img_path) 
                
        # Append the image and its corresponding label to the output
        images.append(image)                
    images = np.array(images, dtype = 'float32')
    
    return images
```

```python
test_images = load_test()
```

```python
print(len(test_images))
```

We generate the vector of prediction probabilities and assign the highest probability for each image as a label to the image.

```python
predictions = model.predict(test_images)     # Vector of probabilities
pred_labels = np.argmax(predictions, axis = 1) # We take the highest probability
```

```python
def display_random_image(class_names, images, labels):
    """
        Display a random image from the images array and its correspond label from the labels array.
    """
    
    index = np.random.randint(images.shape[0])
    plt.figure()
    plt.imshow(images[index].astype('uint8'))
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.title('Image #{} : '.format(index) + class_names[labels[index]])
    plt.show()
```

We display some example images to get an idea on the accuracy of the predictions. As seen below, the predictions are accurate (albeit for just the 25 images shown below). As this is an unsupervised neural network, we cannot test the accuracy of the model, as the test set is not labelled. However, from a manual visual scan of the outputs, the model appears to have performed sufficiently accurately for our later use in testing whether the angle of the scan has a noticeable influence on the prediction of the diagnosis by the diagnosing model.
